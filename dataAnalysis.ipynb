## 0. Set up environment

```python
%matplotlib inline

import numpy as np
import pandas as pd
import scipy.signal as sig
```

## 1. Set up parameters

```python
plt.close("all")

FNameIn = '[Data, StemCell] Pillar1 hESC-CDM3-1A 3p5V All1-ch53.txt';
FHighPass = 1          # high pass filter cutoff [Hz]
FLowPass = 1e3         # low pass filter cutoff [Hz]
highthreshhold = 500   # [uV]
lowthreshhold = 0      # [uV]
rightIndex = 3800      # index for left bound of each action potential
leftIndex = 1500       # index for right bound of each action potential
Irmp = 700             # index from left for resting potential
startOverlayNum = 10
numOfOverlay = 30

data = pd.read_csv(FNameIn, header=3, delimiter='\t')
```

## 2. High pass and low pass filter data and plot

```python
highPassData = []
filteredData = []
timeData = np.array(data.iloc[:,0])
deltaT = data.iloc[1,0] - data.iloc[0,0]  # [ms]
Fs = (1/deltaT*1e3)*2                     # Nyquist rate [Hz]
b, a = sig.butter(4, FHighPass/Fs, 'highpass')
highPassData = sig.filtfilt(b, a, data.iloc[:,1])
b, a = sig.butter(4, FLowPass/Fs, 'lowpass')
filteredData = sig.filtfilt(b, a, highPassData)

### Plot filtered data
plt.figure(1)
plt.plot(timeData/1e3, filteredData/1e3)
plt.xlabel('Time [s]',fontsize=14)
plt.ylabel('Amplitude [mV]',fontsize=14)
plt.ylim(-10,10)
```

## 3. Analyze peaks

```python
peakMax = highthreshhold;
peakTime = 0;
index = 0;
peakID = 0;
peaks = np.zeros((1000, 10)) # later remove all the zeros in this array

for i in range(len(filteredData)):
    if filteredData[i] > peakMax:
        index = 1
        peakMax = filteredData[i]
        peakTime = timeData[i]
    if (index > 0) and (filteredData[i] < lowthreshhold):
        index = 0
        peakID += 1
        peaks[peakID,0] = peakTime     # peak time
        peaks[peakID,1] = peakMax      # peak maximum voltage
        peakMax = highthreshhold
      
for i in range(1,peakID-1):
    j = int(peaks[i,0]/deltaT)+1
    peaks[i,2] = filteredData[j-leftIndex+Irmp]    # resting membrane potential
    peaks[i,3] = peaks[i,1] - peaks[i,2]             # peak amplitude
    RP50Voltage = peaks[i,1] - peaks[i,3]*0.5
    RP90Voltage = peaks[i,1] - peaks[i,3]*0.9
    I50 = np.argmin(np.absolute(filteredData[j:j+rightIndex]-RP50Voltage))
    I90 = np.argmin(np.absolute(filteredData[j:j+rightIndex]-RP90Voltage))
    I50 = I50 + j - 1
    I90 = I90 + j - 1
    I3 = j-leftIndex+Irmp
    peaks[i,4] = timeData[I50]  # RP50Time
    peaks[i,5] = timeData[I90]  # RP90Time
```

## 4. Plot last action potential for checking

```python
plt.figure(2)
plt.plot(timeData[j-leftIndex:j+rightIndex]/1e3, filteredData[j-leftIndex:j+rightIndex]/1e3)
plt.plot(timeData[I50]/1e3, filteredData[I50]/1e3,'r.',markersize=10)     # APD left
plt.plot(timeData[I90]/1e3, filteredData[I90]/1e3,'r.',markersize=10)     # APD right
plt.plot(timeData[I3]/1e3, filteredData[I3]/1e3,'g.',markersize=10)       # min
plt.plot(timeData[j]/1e3, filteredData[j]/1e3,'k.',markersize=10)         # max
plt.xlabel('Time [s]',fontsize=14)
plt.ylabel('Voltage [mV]',fontsize=14)
plt.show()
```

## 5. Calculate APD50, APD90, and APD90/APD50

```python
peaks[:,6] = peaks[:,4] - peaks[:,0]
peaks[:,7] = peaks[:,5] - peaks[:,0]
peaks[:,8] = peaks[:,7]/peaks[:,6]


### Calculate beat interval
for i in range(1,peakID):
    peaks[i,9] = peaks[i,0] - peaks[i-1,0]
peaks[:,9] /= 1e3


### Remove all zeros
peaks = peaks[:peakID,:]
```

## 6. Plotting results

```python
### Plot action potential amplitude vs time
plt.figure(11)
plt.plot(peaks[:,0]/1e3, peaks[:,3]/1e3,'.')
plt.xlabel('Time [s]',fontsize=14)
plt.ylabel('Amplitude [mV]',fontsize=14)
plt.title('Action Potential Amplitude vs. Time',fontsize=14)
plt.ylim(0,5)
plt.show()

### Plot APD vs time
plt.figure(12)
plt.plot(peaks[:,0]/1e3, peaks[:,6],'.');
plt.xlabel('Time [s]',fontsize=14)
plt.ylabel('APD50 [ms]',fontsize=14)
plt.title('Action Potential Duration at 50% Repolarization vs. Time',fontsize=14)
plt.ylim(0,500)
plt.show()

### Plot peak intervals vs time
plt.figure(13)
plt.plot(peaks[:,0]/1e3, peaks[:,9],'.');
plt.xlabel('Time [s]',fontsize=14)
plt.ylabel('Beating Interval [s]',fontsize=14)
plt.title('Beating Interval vs. Time',fontsize=14)
plt.ylim(0,3)
plt.show()
```

## 7. Overlap action potentials [Optional]

```python
tmp = np.zeros((len(filteredData[j-leftIndex:j+rightIndex]),numOfOverlay))
plt.figure(21)
for i in range(startOverlayNum,startOverlayNum+numOfOverlay):
    j = int(peaks[i,0]/deltaT)+1
    plt.plot((timeData[j-leftIndex:j+rightIndex] - peaks[i,0])/1e3, filteredData[j-leftIndex:j+rightIndex]/1e3)
    tmp[:,i-startOverlayNum] = filteredData[j-leftIndex:j+rightIndex]
plt.xlabel('Time [s]',fontsize=14)
plt.ylabel('Amplitude [mV]',fontsize=14)  

    
offsetTimeData = (timeData[j-leftIndex:j+rightIndex]-peaks[i,0])/1e3; # [s]
averagedChannelData = np.mean(tmp, axis=1)/1e3;          # [mV]
Vmax = np.amax(averagedChannelData)            # max
Imax = np.argmax(averagedChannelData) 
Vmin = np.amin(averagedChannelData)            # min
Imin = np.argmin(averagedChannelData)
Vamp = Vmax - averagedChannelData[Irmp]        # AP amplitude [mV] 
Vhyper = averagedChannelData[Irmp] - Vmin      # hyperpolarization amplitude [mV] 
RP50Voltage = Vmax - Vamp*0.5;
RP90Voltage = Vmax - Vamp*0.9;
I50 = np.argmin(np.absolute(averagedChannelData[Imax:]-RP50Voltage))    # index with respect to Imax
I90 = np.argmin(np.absolute(averagedChannelData[Imax:]-RP90Voltage))    # index with respect to Imax
APD50 = I50*deltaT                             # APD50 [ms] 
APD90 = I90*deltaT                             # APD90 [ms] 
APDratio = APD90/APD50
Vratio = Vhyper/Vamp
DepoloarizationTime = (Imax-Irmp)*deltaT       # depolarization time [ms]
DepoloarizationVel = Vamp/DepoloarizationTime  # depolarization velocity [mV/ms]
APD = (Imin-Irmp)*deltaT                       # action potential duration [ms]
```

## 8. Plot parameters for the overlapped action potential [Optional]

```python
plt.figure(22)
plt.plot(offsetTimeData, averagedChannelData, 'b', linewidth=1.1)
plt.plot(offsetTimeData[Imax], averagedChannelData[Imax],'k.', markersize=10)     # max
plt.text(offsetTimeData[Imax], averagedChannelData[Imax],' max')
plt.plot(offsetTimeData[Imin], averagedChannelData[Imin],'k.', markersize=10)    # min
plt.text(offsetTimeData[Imin], averagedChannelData[Imin],' min') 
plt.plot(offsetTimeData[Irmp], averagedChannelData[Irmp],'g.', markersize=10)     # resting potential
plt.text(offsetTimeData[Irmp], averagedChannelData[Irmp],' resting')  
plt.plot(offsetTimeData[Imax+I50], averagedChannelData[Imax+I50],'r.', markersize=10)          # APD50
plt.text(offsetTimeData[Imax+I50], averagedChannelData[Imax+I50],' APD50')
plt.plot(offsetTimeData[Imax+I90], averagedChannelData[Imax+I90],'r.', markersize=10)          # APD90
plt.text(offsetTimeData[Imax+I90], averagedChannelData[Imax+I90],' APD90')   
plt.xlabel('Time [s]',fontsize=14)
plt.ylabel('Amplitude [mV]',fontsize=14)
```

## 9. Output data
